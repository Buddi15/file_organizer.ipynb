```python
import os
import shutil
```

```python
FILE_TYPES = {
    'Documents': ['.pdf', '.docx', '.txt', '.doc', '.xlsx', '.pptx'],
    'Images': ['.jpg', '.jpeg', '.png', '.gif', '.bmp'],
    'Videos': ['.mp4', '.mkv', '.avi', '.mov']
}
```

```python
def get_category(extension):
    for category, extensions in FILE_TYPES.items():
        if extension.lower() in extensions:
            return category
    return 'Others'
```

```python
path = r"C:\Users\gondi\OneDrive"
```

```python
path = r"C:\Users\gondi\Downloads"
```

```python
for filename in os.listdir(path):
    file_path = os.path.join(path, filename)

    if not os.path.isfile(file_path):
        continue

    if filename.lower() == "desktop.ini":
        continue

    _, ext = os.path.splitext(filename)
    folder = get_category(ext)

    folder_path = os.path.join(path, folder)
    os.makedirs(folder_path, exist_ok=True)

    try:
        shutil.move(file_path, os.path.join(folder_path, filename))
        print(f"Moved: {filename} -> {folder}/")
    except Exception as e:
        print(f"Could not move {filename}: {e}")
```

```python
Moved: resume.pdf -> Documents/
Moved: photo.jpg -> Images/
Moved: movie.mp4 -> Videos/
Moved: notes.txt -> Documents/
Moved: vacation.png -> Images/
Moved: music.mp3 -> Others/
```
